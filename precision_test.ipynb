```python
import matplotlib.pyplot as plt
# from matplotlib.patches import Polygon
plt.rcParams["figure.figsize"] = (5,5)
from shapely import Polygon
import numpy as np
np.random.seed(33)


x = [(np.random.random()*10) for i in range(5)]
y = [(np.random.random()*10) for i in range(5)]

p = zip(x,y)
```

```python
x = list(p)
print(x)
```

```python
Poly = Polygon(x)
Poly.exterior
```

```python
xp,yp = Poly.exterior.xy
plt.plot(xp,yp)
plt.show()
```

```python
# Check if a point is inside a polygon or not
x = np.arange(0, 9, 0.3)
y = np.arange(0, 9, 0.3)
print(x,y)
```

```python
import numpy as np
x = (1,2,43,5,65)
y = np.array(x)
print(y)
```

```python
from shapely.geometry import Point
xps, yps = [], []
for xi in x:
    for yi in y:
        p = Point(xi,yi)
        if Poly.contains(p):
            xps += [p.x]
            yps += [p.y]

plt.plot(xp,yp)
plt.scatter(xps,yps, color='r')
```

```python
import gymnasium as gym
# Example of creating and using the environment
env = gym.make('ThreeAgentGridworld-v0', render_mode='human')
env.metadata['render_fps'] = 40
obs, info = env.reset()
env.render()
done = False
while not done:
    action = env.action_space.sample()
    obs, reward, done, _,  info = env.step(action)
    env.render()
    print(f"Obs: {obs}, Reward: {reward}, Done: {done}")
```
